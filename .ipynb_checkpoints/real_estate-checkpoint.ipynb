```python
import numpy as np
import tensorflow as tf
import pandas as pd
import scipy
tf.random.set_seed(42)
np.random.seed(42)

import tensorflow.keras as K
from tensorflow_probability import distributions as tfd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Activation, Concatenate
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_boston
from sklearn.preprocessing import LabelEncoder

import matplotlib as  mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns


def remove_ax_window(ax):
    """
        Remove all axes and tick params in pyplot.
        Input: ax object.
    """
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)    
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)  
    ax.tick_params(axis=u'both', which=u'both',length=0)
    
```

```python
class MDN(tf.keras.Model):

    def __init__(self, neurons=100, components = 2):
        super(MDN, self).__init__(name="MDN")
        self.neurons = neurons
        self.components = components
        
        self.h1 = Dense(neurons, activation="relu", name="h1")
#         self.h2 = Dense(neurons, activation="relu", name="h2")
        
        self.alphas = Dense(components, activation="softmax", name="alphas")
        self.mus = Dense(components, activation="nnelu", name="mus")
        self.sigmas = Dense(components, activation="nnelu", name="sigmas")
        self.pvec = Concatenate(name="pvec")
        
    def call(self, inputs):
        x = self.h1(inputs)
#         x = self.h2(x)
        
        alpha_v = self.alphas(x)
        mu_v = self.mus(x)
        sigma_v = self.sigmas(x)
        
        return self.pvec([alpha_v, mu_v, sigma_v])
    
```

```python
def nnelu(input):
    """ Computes the Non-Negative Exponential Linear Unit
    """
    return tf.add(tf.constant(1, dtype=tf.float32), tf.nn.elu(input))

def nnelu_mu(input):
    """ Computes the Non-Negative Exponential Linear Unit
    """
    return tf.add(tf.constant(0.99, dtype=tf.float32), tf.nn.elu(input))

def slice_parameter_vectors(parameter_vector):
    """ Returns an unpacked list of paramter vectors.
    """
    return [parameter_vector[:,i*components:(i+1)*components] for i in range(no_parameters)]

def gnll_loss(y, parameter_vector):
    """ Computes the mean negative log-likelihood loss of y given the mixture parameters.
    """
    alpha, mu, sigma = slice_parameter_vectors(parameter_vector) # Unpack parameter vectors
    
    gm = tfd.MixtureSameFamily(
        mixture_distribution=tfd.Categorical(probs=alpha),
        components_distribution=tfd.Normal(
            loc=mu,       
            scale=sigma))
    
    log_likelihood = gm.log_prob(tf.transpose(y)) # Evaluate log-probability of y
    
    return -tf.reduce_mean(log_likelihood, axis=-1)

tf.keras.utils.get_custom_objects().update({'nnelu': Activation(nnelu)})
tf.keras.utils.get_custom_objects().update({'nnelu_mu': Activation(nnelu_mu)})
```

```python
no_parameters = 3
components = 4
neurons = 100

opt = tf.optimizers.Adam(1e-5)

tensorboard = TensorBoard(log_dir='tbdir/', histogram_freq=0, write_graph=True, write_images=False)
mon = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=0, mode='auto')
```

```python
mdn = MDN(neurons=neurons, components=components)
mdn.compile(loss=gnll_loss, optimizer=opt)
```

```python
def plot_rate(ax, index,color_index):
    alpha, mu, sigma = slice_parameter_vectors(mdn_4.predict(x_test[index].reshape(1,-1)))

    gm = tfd.MixtureSameFamily(
            mixture_distribution=tfd.Categorical(probs=alpha),
            components_distribution=tfd.Normal(
                loc=mu,       
                scale=sigma))
    pyx = gm.prob(x)
    
    ax.plot(x,pyx,alpha=1, color=sns.color_palette()[color_index], linewidth=2, label="PDF for prediction {}".format(index))
    
    
x = np.linspace(-400,1000,int(1e3))

plt.figure(figsize=(10,7))
ax = plt.gca()

plot_rate(ax, 18,0)
plot_rate(ax, 600,1)
# plot_rate(ax, 60,2)
# plot_rate(ax, 100,3)

ax.set_xlabel("X10Y Value")
ax.set_ylabel("p(X10Y Value)")

ax.legend(loc=0, borderaxespad=0.1, framealpha=1.0, fancybox=True, ncol=1, shadow=True, frameon=False, fontsize=22)
plt.title('Mixture Density Network')
plt.tight_layout()
plt.show()
```
