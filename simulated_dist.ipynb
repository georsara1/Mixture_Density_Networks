```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import numpy as np
sns.set()
```

```python
atm_tourist = random.normal(loc=300000, scale=50000, size=(1, 300000))
atm_non_tourist = random.normal(loc=180000, scale=30000, size=(1, 200000))
```

```python

plt.figure(figsize = (8,4))
sns.distplot(atm_tourist, hist = False, label = 'Tourist ATMs')
sns.distplot(atm_non_tourist, hist = False, label = 'Non Tourist ATMs')
plt.savefig('atm_hist.jpg')
plt.show()
```

```python
xx = np.linspace(100000,400000,int(1e3))

pdf1 = scipy.stats.norm.pdf(xx,300000,20000)
pdf2 = scipy.stats.norm.pdf(xx,240000,15000)
```

```python
plt.plot(xx,pdf1, label = 'Touristic ATMs')
plt.plot(xx,pdf2, label = 'Non Touristic ATMs')
plt.legend()
plt.savefig('dist1_and_2.jpg')
plt.show()
```

```python
pdf_sum = 0.3*pdf1+0.7*pdf2

plt.plot(xx,pdf1, label = 'Touristic ATMs')
plt.plot(xx,pdf2, label = 'Non Touristic ATMs')
plt.plot(xx,pdf_sum, label = 'Mixture')
plt.legend()
plt.savefig('all_dist.jpg')
plt.show()
```
